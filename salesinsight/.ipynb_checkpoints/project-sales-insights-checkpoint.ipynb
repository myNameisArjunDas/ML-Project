```python
#importing important libraries
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
try:
    host = 'localhost'
    user = 'root'
    password = '123Amajeec016!'
    database = 'sales'
    
    # Create an SQLAlchemy engine
    engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}/{database}")

    # Load each table into separate dataframes
    customers_df = pd.read_sql("SELECT * FROM customers", engine)
    date_df = pd.read_sql("SELECT * FROM date", engine)
    markets_df = pd.read_sql("SELECT * FROM markets", engine)
    products_df = pd.read_sql("SELECT * FROM products", engine)
    transactions_df = pd.read_sql("SELECT * FROM transactions", engine)
except Exception as err:
    print("Error:", err)
```

```python
#Creating dictionary of all dataframes
dfs = {
    'customers': customers_df,
    'date': date_df,
    'markets': markets_df,
    'products': products_df,
    'transactions': transactions_df
}

# Iterate over the dictionary and check for null values in each dataframe
for name, df in dfs.items():
    print(f"Null values in DataFrame '{name}':")
    null_columns = df.isnull().sum()
    print(null_columns)
    print("\n")
```

```python
# Iterate over the dictionary and check for duplicate values in each dataframe
for name, df in dfs.items():
    print(f"Duplicate values in DataFrame '{name}':")
    duplicates = df[df.duplicated()]
    if not duplicates.empty:
        print(duplicates)
    else:
        print("No duplicates found.")
    print("\n")
```

```python
#checking the column information about the dataframe
for name, df in dfs.items():
    print(f"Information of DataFrame '{name}':")
    df_info = df.info()
    print(df_info)
    print("\n")
```

```python
#checking for unique values and calculate the sum of unique values in the column in the dataframe
for name, df in dfs.items():
    print(f"Sum of unique values in DataFrame '{name}':")
    for column in df.columns:
        sum_unique_values = df[column].nunique()
        print(f"Column '{column}': {sum_unique_values}")
    print("\n")
```
